```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/jarif87/DataSets/main/house-prices.csv'

response = requests.get(url)

data = pd.read_csv(url)

data.head()
```

```python
plt.plot(data["Price"])
plt.show()
```

```python
plt.boxplot(data["Price"])
plt.show()
```

```python
plt.hist(data["Price"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data["Price"],data["SqFt"])
plt.xlabel("Price")
plt.ylabel("Sqft")
plt.title("Scatter Plot")
plt.show()
```

```python
my_dict={"a":10,"b":20,"c":30,"d":40,"e":50,"f":60}

x=my_dict.keys()
y=my_dict.values()
plt.figure(figsize=(10,10))
plt.bar(x,y)
plt.show()
```
