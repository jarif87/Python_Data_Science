```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/jarif87/DataSets/main/house-prices.csv'

response = requests.get(url)

data = pd.read_csv(url)

data.head()
```

```python
data.columns
```

```python
data.plot(figsize=(10,10),legend=True)
plt.show()
```

```python
data["Price"].plot(figsize=(10,10),legend=True)
plt.show()
```

```python
data.plot(x="Price",y="Bedrooms",kind="scatter",figsize=(10,10))
plt.show()
```

```python
data.plot(title="Pandas Plot",figsize=(10,10))
plt.show()
```

```python
data["SqFt"].plot(kind="box",title="Pandas Boxplot",figsize=(10,10))
plt.show()
```

```python
data["SqFt"].plot(kind="hist",title="Pandas Boxplot\n",figsize=(10,10))
plt.show()
```
