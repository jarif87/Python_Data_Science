```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/jarif87/DataSets/main/ADANIPORTS.NS.csv'

response = requests.get(url)

data = pd.read_csv(url)

data.head()
```

```python
data["H-L"] = data.High - data.Low
data["100MA"] = data["Close"].rolling(100).mean()

plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
ax.scatter(data.index, data["H-L"], data["100MA"])
ax.set_xlabel("Index")
ax.set_ylabel("Data H-L")
ax.set_zlabel("Data 100 MA", fontsize=15, labelpad=2)
plt.show()
```
