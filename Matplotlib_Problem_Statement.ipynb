* you have been provided with dataset that lists ohio state's leading causes of death from the year 2012

* using two data points cause of death and percentile,draw a pie chart to visualize the dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause=["chronic disease","injury","influenja","alzhemier","sepsis","others"]
perceptile=[62,20,5,30,26,45]
```

```python
plt.figure(figsize=(10,10))
explode=(0,0,0,0.2,0,0)
plt.pie(perceptile,labels=cause,autopct="%1.1f%%",startangle=90,explode=explode)
plt.axis("equal")
plt.title("Pie Chart")
plt.show()
```
