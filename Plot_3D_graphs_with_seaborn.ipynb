```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/jarif87/DataSets/main/ADANIPORTS.NS.csv'

response = requests.get(url)

data = pd.read_csv(url)

data.head()
```

```python
data["H-L"]=data.High-data.Low
data["100MA"]=data["Close"].rolling(100).mean()
```

```python
plt.figure(figsize=(10,10))
sns.set_style("darkgrid")
ax=plt.axes(projection="3d")
ax.scatter(data.index,data["H-L"],data["100MA"])
ax.set_xlabel("index")
ax.set_ylabel("H-L")
ax.set_zlabel("100MA")
ax.zaxis.label.set_size(10)
ax.dist = 12

plt.show()
```

```python
plt.figure(figsize=(10,10))
x1=np.linspace(0,10,100)
y1=np.cos(4*x1)
z1=np.sin(4*x1)
ax=plt.axes(projection="3d")

ax.set_xlabel("Y1")
ax.set_ylabel("Z1")
ax.set_zlabel("X1")
ax.plot3D(y1,z1,x1)
plt.show()
```

```python
def return_z(x,y):
  return 50-(x**2+y**2)
```

```python
plt.figure(figsize=(10,10))
sns.set_style("darkgrid")
x1=np.linspace(-5,5,50)
y1=np.linspace(-5,5,50)
x1,y1=np.meshgrid(x1,y1)
z1=return_z(x1,y1)
ax=plt.axes(projection="3d")
ax.plot_surface(x1,y1,z1)
plt.show()
```
